```python
import numpy as np
import pandas as pd

# Step 9.1 – function to add engineered features

def add_engineered_features(df):
    df = df.copy()
    
    # 1) Smoking burden – pack years
    # If missing, treat as 0 for the product
    if "SMOKYRS" in df.columns and "PACKSPER" in df.columns:
        years = df["SMOKYRS"].fillna(0).clip(lower=0)
        packs = df["PACKSPER"].fillna(0).clip(lower=0)
        df["pack_years"] = years * packs
    
    # 2) Cardio burden – count of major vascular conditions
    cardio_cols = [
        "CVHATT", "HATTMULT", "CBSTROKE", "STROKMUL", "CBTIA",
        "DIABETES", "HYPERTEN", "HYPERCHO", "CVCHF", "CVANGINA", "CVHVALVE",
    ]
    present = [c for c in cardio_cols if c in df.columns]
    if present:
        # treat code '1' as "yes", everything else as 0
        cardio_binary = [(df[c] == 1).astype(float) for c in present]
        df["cardio_burden"] = np.vstack(cardio_binary).sum(axis=0)
    
    # 3) Psych burden – count of psychiatric conditions
    psych_cols = [
        "DEP2YRS", "DEPOTHR", "ANXIETY", "BIPOLAR",
        "SCHIZ", "PTSD", "ALCOHOL", "ABUSOTHR",
    ]
    present_psy = [c for c in psych_cols if c in df.columns]
    if present_psy:
        psych_binary = [(df[c] == 1).astype(float) for c in present_psy]
        df["psych_burden"] = np.vstack(psych_binary).sum(axis=0)
    
    # 4) Sleep burden – number of sleep problems
    sleep_cols = ["APNEA", "INSOMN", "OTHSLEEP"]
    present_sleep = [c for c in sleep_cols if c in df.columns]
    if present_sleep:
        sleep_binary = [(df[c] == 1).astype(float) for c in present_sleep]
        df["sleep_burden"] = np.vstack(sleep_binary).sum(axis=0)
    
    # 5) Any parent dementia / cognitive problems
    if "NACCMOM" in df.columns and "NACCDAD" in df.columns:
        mom = (df["NACCMOM"] == 1).astype(float)
        dad = (df["NACCDAD"] == 1).astype(float)
        df["parent_dementia_any"] = ((mom + dad) > 0).astype(float)
    
    # 6) BMI category from NACCBMI
    def bmi_to_cat(bmi):
        if pd.isna(bmi):
            return np.nan
        if bmi < 18.5:
            return "underweight"
        elif bmi < 25:
            return "normal"
        elif bmi < 30:
            return "overweight"
        else:
            return "obese"
    
    if "NACCBMI" in df.columns:
        df["BMI_CAT"] = df["NACCBMI"].apply(bmi_to_cat)
    
    return df
```

```python
# Step 9.2 – apply feature engineering

train_fe = add_engineered_features(train_df)
val_fe   = add_engineered_features(val_df)
test_fe  = add_engineered_features(test_df)

train_fe.shape, val_fe.shape, test_fe.shape
```

```python
# Step 9.3 – new feature lists including engineered features

id_col = "NACCID"
target_col = "DEMENTED"

feature_cols_fe = [c for c in train_fe.columns if c not in [id_col, target_col]]

# Same base categorical candidates as before, plus BMI_CAT
categorical_candidates_fe = [
    "SEX",
    "HISPANIC",
    "RACE",
    "MARISTAT",
    "NACCLIVS",
    "RESIDENC",
    "PRIMLANG",
    "HANDED",
    "BMI_CAT",
]

categorical_features_fe = [c for c in categorical_candidates_fe if c in feature_cols_fe]
numeric_features_fe = [c for c in feature_cols_fe if c not in categorical_features_fe]

print("Categorical (FE):", categorical_features_fe)
print("Numeric (FE, first 10):", numeric_features_fe[:10])
print("Total categorical (FE):", len(categorical_features_fe), "Total numeric (FE):", len(numeric_features_fe))
```

```python
# Step 9.4 – new preprocessor + logistic regression with engineered features

numeric_transformer_fe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
])

categorical_transformer_fe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore")),
])

preprocessor_fe = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer_fe, numeric_features_fe),
        ("cat", categorical_transformer_fe, categorical_features_fe),
    ]
)

log_reg_fe = Pipeline(steps=[
    ("preprocessor", preprocessor_fe),
    ("model", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",
        solver="lbfgs"
    )),
])

log_reg_fe
```

```python
# Step 9.5 – fit FE model
log_reg_fe.fit(
    train_fe[feature_cols_fe],
    train_fe[target_col]
)

# Evaluate on val set
val_proba_fe = log_reg_fe.predict_proba(val_fe[feature_cols_fe])[:, 1]
val_pred_fe = (val_proba_fe >= 0.5).astype(int)

val_accuracy_fe = accuracy_score(y_val, val_pred_fe)
val_auc_fe = roc_auc_score(y_val, val_proba_fe)
cm_fe = confusion_matrix(y_val, val_pred_fe)

print("FE model – Validation accuracy:", val_accuracy_fe)
print("FE model – Validation ROC-AUC:", val_auc_fe)
print("Confusion matrix:\n", cm_fe)
print("\nClassification report:\n", classification_report(y_val, val_pred_fe))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Step 10.2 – preprocessor for tree-based models (no scaling)

numeric_transformer_tree = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
])

categorical_transformer_tree = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore")),
])

preprocessor_tree = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer_tree, numeric_features_fe),
        ("cat", categorical_transformer_tree, categorical_features_fe),
    ]
)

preprocessor_tree
```

```python
# Step 10.3 – Random Forest pipeline + RandomizedSearchCV

rf_clf = RandomForestClassifier(
    n_estimators=200,
    class_weight="balanced",   # help with class imbalance
    random_state=42,
    n_jobs=-1                  # use all CPU cores
)

rf_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor_tree),
    ("model", rf_clf),
])

# Hyperparameter search space (small, so it runs on laptop)
rf_param_dist = {
    "model__n_estimators": [200, 400],
    "model__max_depth": [None, 8, 12, 16],
    "model__min_samples_split": [2, 5, 10],
    "model__min_samples_leaf": [1, 2, 4],
    "model__max_features": ["sqrt", "log2"],
}

rf_search = RandomizedSearchCV(
    rf_pipeline,
    param_distributions=rf_param_dist,
    n_iter=15,              # 15 random combos
    scoring="roc_auc",
    cv=3,
    verbose=1,
    n_jobs=-1,
    random_state=42,
)

rf_search
```

```python
rf_search.fit(
    train_fe[feature_cols_fe],
    train_fe[target_col]
)
```

```python
print("RF best ROC-AUC (CV):", rf_search.best_score_)
print("RF best params:", rf_search.best_params_)
```

```python
# Step 14.2 – wrapper to predict risk for a single person

def predict_dementia_risk(person_features: dict):
    """
    person_features: dict mapping feature name -> value
      Example keys:
        "NACCAGE", "EDUC", "SEX", "HISPANIC", "RACE",
        "NACCLIVS", "RESIDENC", "TOBAC30", "DIABETES",
        "HYPERTEN", "CBSTROKE", "NACCBMI", "DEP2YRS", etc.
      Any missing features will be treated as NaN and imputed.
    """
    # start with all features as NaN
    row = {col: np.nan for col in feature_cols_fe}
    
    # update with whatever the user provided
    for key, value in person_features.items():
        if key in row:
            row[key] = value
        else:
            # just ignore unknown keys
            pass
    
    X_person = pd.DataFrame([row])
    
    # use the full pipeline (preprocessor + XGBoost)
    prob = xgb_best.predict_proba(X_person)[0, 1]
    cat = risk_category(prob)
    
    return prob, cat
```
